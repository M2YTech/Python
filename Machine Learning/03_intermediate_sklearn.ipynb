# intermediate Machine Learning with SkLearn

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#linear regression
from sklearn.linear_model import LinearRegression

#load dataset
df = sns.load_dataset('tips')

# select feature column from df
x =df[['total_bill']]
# select target column from df
y =df[['tip']]

# load/call linear regression model
model = LinearRegression()

#fit the model means to train the model
model.fit(x,y)

#lets predict the tip for unknown total_bill
model.predict([[50]])

```
